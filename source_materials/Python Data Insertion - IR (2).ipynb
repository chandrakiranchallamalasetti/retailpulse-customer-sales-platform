```python
import pyodbc
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

```python
# Connect to SQL Server
conn_str = (
    "DRIVER={SQL Server};"
    "SERVER=YourServerName;"
    "DATABASE=YourDatabaseName;"
    "Trusted_Connection=yes;"  
    "UID=YourUsername;"
    "PWD=YourPassword;"
)

try:
    
    conn = pyodbc.connect(conn_str)
    cursor = conn.cursor()
    print("Connected to SQL Server successfully!")
    
```

```python
# Function to insert data into SalesPerson table
    def insert_sales_persons():
        first_names = [
            'William', 'James', 'Charles', 'George', 'Richard', 'Joseph',
            'Edward', 'Mary', 'Patricia', 'Linda', 'Barbara', 'Elizabeth',
            'Susan', 'Nancy', 'Karen', 'Helen', 'Sandra', 'Donna', 'Carol',
            'Ruth', 'Sharon', 'Michelle', 'Laura', 'Sarah', 'Kimberly'
        ]
        
        last_names = [
            'Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 'Garcia',
            'Miller', 'Davis', 'Rodriguez', 'Martinez', 'Hernandez', 'Lopez',
            'Gonzalez', 'Wilson', 'Anderson', 'Thomas', 'Taylor', 'Moore',
            'Jackson', 'Martin', 'Lee', 'Perez', 'Thompson', 'White', 'Harris'
        ]
        
      
        sales_persons = []
        for i in range(1, 26): 
            sales_person_id = i + 100
            first_name = first_names[i-1]
            last_name = last_names[i-1]
            sales_persons.append((sales_person_id, first_name, last_name))
        
        
        for sp in sales_persons:
            cursor.execute(
                "INSERT INTO SalesPerson (SalesPersonID, FirstName, LastName) VALUES (?, ?, ?)",
                sp
            )
        
        conn.commit()
        print(f"Inserted {len(sales_persons)} sales persons")
        return sales_persons
```

```python
 # Function to insert data into Customer table
    def insert_customers(sales_persons):
        company_prefixes = [
            'Alpha', 'Beta', 'Gamma', 'Delta', 'Epsilon', 'Zeta', 'Eta', 'Theta',
            'Iota', 'Kappa', 'Lambda', 'Mu', 'Nu', 'Omega', 'Pi', 'Rho', 'Sigma',
            'Tau', 'Upsilon', 'Phi', 'Chi', 'Psi', 'Global', 'National', 'Regional'
        ]
        
        company_suffixes = [
            'Technologies', 'Solutions', 'Systems', 'Enterprises', 'Corporation',
            'Inc.', 'LLC', 'Group', 'Innovations', 'Industries', 'Labs', 'Partners',
            'Consulting', 'Services', 'Associates', 'Networks', 'International',
            'Software', 'Hardware', 'Dynamics', 'Ventures', 'Capital', 'Development',
            'Manufacturing', 'Research'
        ]
        
        states = [
            'AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'IA',
            'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME', 'MI', 'MN', 'MO',
            'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 'NY', 'OH', 'OK',
            'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI',
            'WV', 'WY'
        ]
        
        # Generate the customer data
        customers = []
        for i in range(1, 26):  # 25 customers
            customer_id = i + 200  
            name = f"{company_prefixes[i-1]} {company_suffixes[i-1]}"
            state = random.choice(states)
            sales_person_id = sales_persons[i-1][0] 
            customers.append((customer_id, name, state, sales_person_id))
        
        # Insert data into Customer table
        for customer in customers:
            cursor.execute(
                "INSERT INTO Customer (CustomerID, Name, State, SalesPersonID) VALUES (?, ?, ?, ?)",
                customer
            )
        
        conn.commit()
        print(f"Inserted {len(customers)} customers")
        return customers
```

```python
    # Function to insert data into OrderHeader table
    def insert_order_headers(customers):
        order_headers = []
        start_date = datetime(2025, 1, 1)
        
        for i in range(1, 26):  # 25 orders
            order_number = i + 2000  
            order_date = start_date + timedelta(days=i * 3)
            customer_id = customers[i-1][0]  
            order_headers.append((order_number, order_date, customer_id))
        
        for order in order_headers:
            cursor.execute(
                "INSERT INTO OrderHeader (OrderNumber, OrderDate, CustomerID) VALUES (?, ?, ?)",
                order
            )
        
        conn.commit()
        print(f"Inserted {len(order_headers)} order headers")
        return order_headers
```

```python

    
    # Function to insert data into OrderDetail table
    def insert_order_details(order_headers):
        products = [
            ('Desktop Computer', 1200.00),
            ('Laptop', 1000.00),
            ('Tablet', 500.00),
            ('Smartphone', 800.00),
            ('Monitor', 300.00),
            ('Keyboard', 50.00),
            ('Mouse', 30.00),
            ('Printer', 200.00),
            ('Scanner', 150.00),
            ('External Hard Drive', 120.00),
            ('USB Drive', 20.00),
            ('Webcam', 80.00),
            ('Headphones', 100.00),
            ('Speakers', 70.00),
            ('Router', 90.00),
            ('Modem', 60.00),
            ('Network Switch', 40.00),
            ('Server', 2000.00),
            ('UPS', 250.00),
            ('Software License', 500.00),
            ('Office Chair', 200.00),
            ('Computer Desk', 300.00),
            ('Docking Station', 150.00),
            ('Projector', 600.00),
            ('Smart TV', 700.00)
        ]
        
        order_details = []
        detail_id = 20001  
        
        for order in order_headers:
            order_number = order[0]
            # Each order will have 1-3 items
            num_items = random.randint(1, 3)
            
            selected_products = random.sample(products, num_items)
            
            for product in selected_products:
                item_name = product[0]
                price = product[1]
               
                quantity = 1 if random.random() < 0.7 else random.randint(2, 5)
                
                order_details.append((detail_id, order_number, item_name, quantity, price))
                detail_id += 1
        
       
        for detail in order_details:
            cursor.execute(
                "INSERT INTO OrderDetail (DetailID, OrderNumber, ItemName, Quantity, Price) VALUES (?, ?, ?, ?, ?)",
                detail
            )
        
        conn.commit()
        print(f"Inserted {len(order_details)} order details")
        return order_details
    

    sales_persons_data = insert_sales_persons()
    customers_data = insert_customers(sales_persons_data)
    order_headers_data = insert_order_headers(customers_data)
    order_details_data = insert_order_details(order_headers_data)
    
    print("All data inserted successfully!")
    
except Exception as e:
    print(f"Error: {str(e)}")
    
finally:
   
    if 'conn' in locals():
        conn.close()
        print("Connection closed.")
```
